## Observations and Insights

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import scipy as sp



# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
mouse_metadata.head()
```

```python
study_results.head()
```

```python
# Combine the data into a single dataset
combined_data=pd.merge(mouse_metadata,study_results,how="left",on="Mouse ID")
#displaying the data after combining
combined_data.head()
```

```python
#listing the objects and treatments
mice = combined_data["Mouse ID"].unique()
drugs = combined_data["Drug Regimen"].unique()
time = combined_data["Timepoint"].unique()

# counts of subjects and treatments
num_mouse = len(mice)
num_drugs = len(drugs)
num_measurements = len(time)
num_samples = num_drugs * num_measurements

# summarise in a dataframe
data_summary = pd.DataFrame({"Number of Mice": [num_mouse],
                         "Number of Drug Regimen Treatments": [num_drugs],
                         "Number of Time Measurements": [num_measurements],
                         "Number of Samples": [num_samples]})
data_summary
```

## Summary statistics

```python
# Generate a summary statistics table of mean,
#median, variance, standard deviation,
#and SEM of the tumor volume for each regimen

#tv=tumor volume
mean_tv = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)'],2))
mean_tv  =mean_tv .rename(columns={"Tumor Volume (mm3)":"Mean"})
mean_tv.head()
```

```python
median_tv = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)'],2))
median_tv =median_tv.rename(columns={"Tumor Volume (mm3)":"Median"})
median_tv.head()
```

```python
sem_tv= pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)'],2))
sem_tv =sem_tv.rename(columns={"Tumor Volume (mm3)":"SEM"})
sem_tv.head()
```

```python
variance_tv = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)'],2))
variance_tv =variance_tv.rename(columns={"Tumor Volume (mm3)":"Variance"})
variance_tv.head()
```

```python
std_tv = pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)'],2))
std_tv = std_tv.rename(columns={"Tumor Volume (mm3)":"Std"})
std_tv .head()
```

```python
summary_stat=[mean_tv,median_tv ,sem_tv,variance_tv,std_tv]
stat_result= pd.concat(summary_stat,axis=1,join='inner')
stat_result
```

```python
stat_analysis= pd.DataFrame(round(combined_data.groupby(['Drug Regimen']).describe()['Tumor Volume (mm3)'],2))

stat_analysis
```

## Quartiles, outliers and boxplots

```python
#Maing Question: Calculate the final tumor volume of each mouse across four of the most promising treatment regimens

#first step: finding the top highest tumer volume
highest_tumor_vol =combined_data.groupby(["Mouse ID"]).max().reset_index()
highest_tumor_vol.head()
```

```python
#Calculate the IQR and quantitatively determine if there are any potential outliers.

#merging the original data frame and the highest level of the tumor vol by Mouse ID and timepoint
tumor_data = highest_tumor_vol[['Mouse ID','Timepoint']].merge(combined_data,on=['Mouse ID','Timepoint'],how="outer")

inf_drug= tumor_data.loc[tumor_data["Drug Regimen"] == "Infubinol"]['Tumor Volume (mm3)']
cef_drug = tumor_data.loc[tumor_data["Drug Regimen"] == "Ceftamin"]['Tumor Volume (mm3)']
cap_drug = tumor_data.loc[tumor_data["Drug Regimen"] == "Capomulin"]['Tumor Volume (mm3)']
ramicane_drug = tumor_data.loc[tumor_data["Drug Regimen"] == "Ramicane"]['Tumor Volume (mm3)']
```

```python
#defining the quartile for all top 4 drugs
ramicane_drug_quartiles = ramicane_drug.quantile([.25,.5,.75])
cap_drug_quartiles = cap_drug.quantile([.25,.5,.75])
inf_drug_quartiles = inf_drug.quantile([.25,.5,.75])
cef_drug_quartiles = cef_drug.quantile([.25,.5,.75])

#defining the first/lower quartile for all top 4 drugs
ramicane_drug_lq= ramicane_drug_quartiles[0.25]
cap_drug_lq= cap_drug_quartiles[0.25]
inf_drug_lq = inf_drug_quartiles[0.25]
cef_drug_lq = cef_drug_quartiles[0.25]

#defining the third/upper quartile for all top 4 drugs
ramicane_drug_uq = ramicane_drug_quartiles[0.75]
cap_drug_uq = cap_drug_quartiles[0.75]
inf_drug_uq = inf_drug_quartiles[0.75]
cef_drug_uq = cef_drug_quartiles[0.75]


#Calculating the IQR for all top 4 drugs
ramicane_drug_IQR = ramicane_drug_uq-ramicane_drug_lq
cap_drug_IQR = cap_drug_uq-cap_drug_lq
inf_drug_IQR = inf_drug_uq-inf_drug_lq
cef_drug_IQR = cef_drug_uq-cef_drug_lq


#Calculating the lower Bound for all top 4 drugs
ramicane_drug_LB = ramicane_drug_lq - (1.5*ramicane_drug_IQR)
cap_drug_UB = cap_drug_uq + (1.5*cap_drug_IQR)
inf_drug_LB = inf_drug_lq - (1.5*inf_drug_IQR)
cef_drug_LB = cef_drug_lq- (1.5*cef_drug_IQR)


#Calculating the Upper Bound for all top 4 drugs
ramicane_drug_UB = ramicane_drug_uq + (1.5*ramicane_drug_IQR)
cap_drug_LB = cap_drug_lq - (1.5*cap_drug_IQR)
inf_drug_UB = inf_drug_uq + (1.5*inf_drug_IQR)
cef_drug_UB = cef_drug_uq + (1.5*cef_drug_IQR)


print(f"Outliers Ramicane: {ramicane.loc[(ramicane < ramicane_drug_LB) | (ramicane > ramicane_drug_UB)]}"
      f"Outliers Capomulin:{capomulin.loc[(capomulin < cap_drug_LB) | (capomulin > cap__drug_UB)]}" 
      f"Outliers Infubinol: {infubinol.loc[(infubinol < inf_drug_LB) | (infubinol > inf_drug_UB)]}"
      f"Outliers Ceftamin: {ceftamin.loc[(ceftamin < cef_drug_LB) | (ceftamin > cef_drug_UB)]}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot([ramicane,capomulin,infubinol,ceftamin],
            labels=['Ramicane','Capomulin','Infubinol','Ceftamin'])
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Top 4 Promissing Drug Regimen')
plt.show()
```
